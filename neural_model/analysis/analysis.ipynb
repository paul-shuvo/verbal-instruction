```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {}
for f in os.listdir():
    if '.npy' in f:
        data[f.split('.')[0].upper().replace('_', '-')] = np.load(f, allow_pickle=True)
# a.shape
```

```python
import matplotlib
from random import randint
linestyles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--', '-.', ':']
markers = ['+', '^', 'o', 's', 'p', '*', 'x', '+', '^', 'o', 's', 'p', '*', 'x']


colors = list(matplotlib.colors.cnames.values())
linestyles_ind = [1, 1, 2, 3, 2, 1, 2, 3]
colors_ind = [20, 45, 65, 54, 10, 41, 34, 45]
plt.style.use('fast')
plt.rcParams["font.family"] = "Times New Roman"



fig, ax = plt.subplots()
i = 0
for k, v in data.items():
    ax.plot(v[3][0:15], label=k, linewidth=4, linestyle=linestyles[linestyles_ind[i]] ,color=colors[colors_ind[i]])
    i += 1
ax.set(xlabel='Epoch', ylabel='Accuracy')
fig.suptitle('Validation Accuracy', fontsize=25)

ax.grid()

# fig.savefig("test.png")
# plt.legend(loc="upper left")
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles, labels, bbox_to_anchor=(0.8, -0.15), ncol=2, prop={'size': 16})
# plt.legend(loc='best', prop={'size': fontsize_label})

# fig.subplots_adjust(hspace=0.6)
plt.grid(True)
# plt.grid(which='major')
# plt.grid(which='minor', linestyle = ':')
# plt.draw()
# plt.savefig('val_acc.png', dpi=400)
fig.set_size_inches(10.,6.)
plt.xlabel('Epoch', fontsize=18, color='blue')
plt.ylabel('Accuracy', fontsize=18, color='green')
plt.show()

plt.style.use('default')
```
